```python
import cv2
import os
```

```python
def chop_video_into_images(video_path, output_folder, frame_interval=1):
    # Open the video file
    cap = cv2.VideoCapture(video_path)
    
    # Create the output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Get video information
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))

    # Loop through each frame and save images
    for frame_number in range(0, frame_count, frame_interval):
        # Set the frame position
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)
        
        # Read the frame
        ret, frame = cap.read()

        if not ret:
            break  # Break if there are no more frames

        # Save the frame as an image
        image_path = os.path.join(output_folder, f"frame_{frame_number}.png")
        cv2.imwrite(image_path, frame)

        print(f"Saved frame {frame_number}/{frame_count}")

    # Release the video capture object
    cap.release()
```

```python
# Example usage
video_path = './test_video/example-1.mp4'
output_folder = './test_video/chopped_output_images'
frame_interval = 1  # Save every 10 frames, adjust as needed

chop_video_into_images(video_path, output_folder, frame_interval)
```
