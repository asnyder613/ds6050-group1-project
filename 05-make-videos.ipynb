```python
import os
import imageio

# Path to the directory containing your images
image_folder = '/scratch/ybt7qf/ds6050-group1-project/datasets/eurocitypersonsdataset/images/train-resize'

# Output video file name with MP4 extension
video_name = 'output_video_2.mp4'

# Define the range of image filenames
start_index = 281
end_index = 381

# Get the list of image filenames within the specified range
images = [f"torino_{str(i).zfill(5)}.png" for i in range(start_index, end_index + 1) if os.path.isfile(os.path.join(image_folder, f"torino_{str(i).zfill(5)}.png"))]

# Sort the images based on their filenames
images.sort()

# Create a video writer object with H.264 codec and MP4 format using imageio
video_writer = imageio.get_writer(video_name, fps=1, format='FFMPEG', codec='libx264')

# Iterate through the images and write them to the video
for image in images:
    img_path = os.path.join(image_folder, image)
    frame = imageio.imread(img_path)
    video_writer.append_data(frame)

# Close the video writer
video_writer.close()

print(f"Video {video_name} created successfully.")
```
